#  Image to Pencil Sketch with Python

# Importing Library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

# Reading Image

```python
img=cv2.imread("Original img.jpg")
```

# Reading image in RGB format

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.axis("off")
```

# Converting original Image into gray scale image

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(img_gray)
plt.axis("off")
```

# Inverting Image

```python
img_inverted = 255 -img_gray
plt.imshow(img_inverted)
plt.axis("off")
```

# Blurring the image by using Gaussian Function

```python
blurred_img = cv2.GaussianBlur(img_inverted,(21,21),0)
plt.imshow(blurred_img)
plt.axis("off")
```

# Inverting the blurred image into pencil sketch image 

```python
inverted_blurred = 255 - blurred_img
sketch_img = cv2.divide(img_gray,inverted_blurred,scale=250)


kernal=np.array([[-1,-1,-1],
                 [-1,9,-1],
                 [-1,-1,-1]])

final_sketch=cv2.filter2D(sketch_img,-1,kernal)
plt.axis("off")
plt.imshow(final_sketch,cmap="gray")
```

# Displaying orginal image and pencil sketch image

```python
f,axarr = plt.subplots(1,2)

axarr[0].imshow(img_rgb)
axarr[0].axis("off")
axarr[1].imshow(final_sketch,cmap='gray')
axarr[1].axis("off")
```
