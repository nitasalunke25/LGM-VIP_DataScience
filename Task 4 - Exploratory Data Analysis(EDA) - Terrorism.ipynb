```python
data = df[['Year','Month','day','Country','latitude','longitude',"Attacktype",'Wound','target1',
           'summary','Group','Targettype','motive']]
```

```python
data.head()
```

```python
data['Wound']=data['Wound'].fillna(0)
data['target1']= data['target1'].fillna(0)
```

```python
plt.figure(figsize=(10,20))
cor = df.corr()
plt.show()
```

```python
year = data['Year'].unique()
years_count = data['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (16,11))
sns.barplot(x=year,
           y=years_count,
           palette = "RdBu")
plt.xticks(rotation = 50)
plt.xlabel('Attacking Year',fontsize= 20)
plt.ylabel('Number of Attacks Each Year',fontsize=20)
plt.title('Attacks In Years',fontsize=30)
plt.show()
```

```python
plt.subplots(figsize=(20,10))
sns.barplot(df['country'].value_counts()[:20].index)
plt.title('Top 20 most affected countries', fontsize=25)
plt.ylabel('Number of Attacks',fontsize=20)
plt.xlabel('country',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(43,10))
sns.countplot(df['country'],palette="magma")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(40,10))
sns.countplot(df['Month'],palette='viridis')
plt.title('Number of Terrorist Attacks per Month',fontsize='20')
plt.ylabel("Number of attacks")
plt.xlabel("Month")
plt.xticks(rotation='40')
plt.show()
```

```python
plt.figure(figsize=(10,20))
plt.title('Top 10 Affected Regions',fontsize='30')
df['region'].value_counts().head(30).plot(kind='pie',autopct='%.02f%%')
```

```python
plt.figure(figsize=(40,10))
sns.countplot(df['Year'],palette='viridis')
plt.title('Number of Terrorist Attacks per Year',fontsize='20')
plt.ylabel("Number of attacks")
plt.xlabel("year")
plt.xticks(rotation='40')
plt.show()
```

```python
plt.figure(figsize=(10,20))
plt.title('Top 10 Affected Country',fontsize='30')
df['country'].value_counts().head(30).plot(kind='pie',autopct='%.02f%%')
```
